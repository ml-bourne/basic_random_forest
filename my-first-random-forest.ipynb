```python
import pandas as pd

df = pd.read_csv("train/subj10_series5_data.csv")
```

```python
df.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
df.drop([1], axis=0).head()
```

```python
X_train = df.drop(["id"], axis=1).values
X_train
```

```python
df_events = pd.read_csv("train/subj10_series5_events.csv")
df_events.head()
y_train = df_events.drop(["id"], axis=1).values
y_train
```

```python
clf = RandomForestClassifier(n_jobs=4)
clf.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, roc_auc_score

yp = clf.predict(X_train)
yp
```

```python
accuracy = accuracy_score(y_train, yp)
accuracy
```
